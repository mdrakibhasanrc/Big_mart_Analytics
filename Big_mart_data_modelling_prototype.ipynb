# fact_table

```python
fact_table=df.merge(Item_dim,left_on='Item_Id',right_on='Item_count_id')\
            .merge(Item_detail_dim,left_on='Item_Id',right_on='Item_detail_Id')\
            .merge(outlet_dim,left_on='Item_Id',right_on='outlet_Id')\
            .merge(outlet_detail_dim,left_on='Item_Id',right_on='outlet_detail_Id')\
            .merge(Item_sales_dim,left_on='Item_Id',right_on='Item_sales_Id')
```
